### Airline Delay Data Cleanup

```
import pyspark.sql.functions as f

#Load 2018 and 2019 data into spark dataframes
airline2018 = spark.table("airline2018")
airline2019 = spark.table("airline2019")

#Combine 2018 and 2019 data
airlines_201819 = airline2018.union(airline2019)

#Drop unused columns
columns_to_drop = ['Unnamed: 49']
airlines_201819 = airlines_201819.drop(*columns_to_drop)
display(airlines_201819)
```

#### Save 2018 and 2019 spark dataframe as a table

```
permanent_table_name = "airline_2018_19_final"
airlines_201819.write.mode("overwrite").saveAsTable(permanent_table_name)
```

#### Select columns for final Tableau data analysis

```
#Load 2018 and 2019 data into spark dataframe 
airline_2018_19_final = spark.table("airline_2018_19_final")

#Select columns for Tableau Data Analysis
airline_201819_tableau=airline_2018_19_final.select('YEAR','QUARTER','MONTH','DAY_OF_WEEK','FL_DATE','OP_UNIQUE_CARRIER','ORIGIN','ORIGIN_CITY_NAME','DEST','DEST_CITY_NAME','FLIGHTS','DEP_TIME','DEP_DELAY','DEP_DELAY_GROUP','DEP_DELAY_NEW','ARR_TIME','ARR_DELAY','ARR_DELAY_NEW','ARR_DELAY_GROUP','CANCELLED','CANCELLATION_CODE','DIVERTED','CARRIER_DELAY','WEATHER_DELAY','NAS_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY')
```

#### Modify table to include only top 4 airline carriers

```
#Save Table with airline carriers in ['AA','DL','UA','WN']

airline_201819_tableau=airline_201819_tableau.filter(airline_201819_tableau.OP_UNIQUE_CARRIER.isin(['AA','DL','UA','WN']))
permanent_table_name = "airline_2018_19_tableau"
airline_201819_tableau.write.mode("overwrite").saveAsTable(permanent_table_name)
```

#### Save spark dataframe to table to include 2018,2019 data and 4 top airline carriers

```
airline = spark.table("airline_2018_19_tableau")
display(airline)
```

### Itenerary Fare Data Cleanup

```
#Combine itenerary fare data of 2018,2019

airline2018_01_price = spark.table("2018_01_price")
airline2018_02_price = spark.table("2018_02_price")
airline2018_03_price = spark.table("2018_03_price")
airline2018_04_price = spark.table("2018_04_price")
airline2019_01_price = spark.table("2019_01_price")
airline2019_02_price = spark.table("2019_02_price")

airlines_price = airline2018_01_price.union(airline2018_02_price).union(airline2018_03_price).union(airline2018_04_price).union(airline2019_01_price).union(airline2019_02_price)
```

```
permanent_table_name = "airlines_price"
airline2018_01_price.write.mode("overwrite").saveAsTable(permanent_table_name)

airlines_price=airlines_price.filter(airlines_price.REPORTING_CARRIER.isin(['AA','DL','UA','WN']))
```

#### Drop irrelevant columns

```
from pyspark.sql.types import IntegerType
airlines_price = airlines_price.withColumn("Passengers", airline2018_01_price["Passengers"].cast(IntegerType()))
airlines_price = airlines_price.withColumn("MARKET_FARE", airline2018_01_price["MARKET_FARE"].cast(IntegerType()))

airlines_price=airlines_price.drop('_c11','MARKET_MILES_FLOWN','DEST_STATE_NM','ORIGIN_STATE_ABR','ITIN_ID')
```

```
permanent_table_name = "airline_price_tableau"

airlines_price.write.mode("overwrite").saveAsTable(permanent_table_name)
```

```
airline_price_tableau = spark.table("airline_price_tableau")
display(airline_price_tableau)
```
